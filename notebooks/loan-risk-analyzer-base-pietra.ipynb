## 1. Configuração do Ambiente

```python
"""Carregamento do dataset de risco de crédito (CSV).

Lê o arquivo CSV de empréstimos, valida as colunas esperadas e retorna
uma matriz NumPy de atributos (X) e um vetor alvo (y).
"""

from __future__ import annotations 
import csv 
from dataclasses import dataclass 
from pathlib import Path 
from typing import List 
import numpy as np


# DATASET_PATH = ( Path(__file__).resolve() .parents[2] / "data" / "loan_data.csv" )

DATASET_PATH = Path.cwd().parent / "data" / "loan_data.csv"


SEED = 20
```

## 2. Carregamento do Dataset

```python
def load_csv_as_matrix(path):
    gender_map = {
        "female": 0,
        "male": 1,
    }

    education_map = {
        "High School": 0,
        "Associate": 1,
        "Bachelor": 2,
        "Master": 3,
        "Doctorate": 4,
    }

    home_map = {
        "RENT": 0,
        "OWN": 1,
        "MORTGAGE": 2,
        "OTHER": 3,
    }

    intent_map = {
        "PERSONAL": 0,
        "EDUCATION": 1,
        "MEDICAL": 2,
        "VENTURE": 3,
        "HOMEIMPROVEMENT": 4,
        "DEBTCONSOLIDATION": 5,
    }

    default_map = {
        "No": 0,
        "Yes": 1,
    }

    X = []
    y = []

    with open(path, newline="", encoding="utf-8") as file:
        reader = csv.DictReader(file)

        for row in reader:
            features = [
                float(row["person_age"]),
                gender_map[row["person_gender"]],
                education_map[row["person_education"]],
                float(row["person_income"]),
                float(row["person_emp_exp"]),
                home_map[row["person_home_ownership"]],
                float(row["loan_amnt"]),
                intent_map[row["loan_intent"]],
                float(row["loan_int_rate"]),
                float(row["loan_percent_income"]),
                float(row["cb_person_cred_hist_length"]),
                float(row["credit_score"]),
                default_map[row["previous_loan_defaults_on_file"]],
            ]

            X.append(features)
            y.append(int(row["loan_status"]))

    X = np.array(X, dtype=float)
    y = np.array(y, dtype=int)

    return X, y


X, y = load_csv_as_matrix(DATASET_PATH)

print("X shape:", X.shape)
print("y shape:", y.shape)

print("\nPrimeiras 5 linhas:")
print(X[:5])

print("\nPrimeiros 5 rótulos:")
print(y[:5])
```
